```python
import functools

def timethis(func):

    @functools.wraps(func)
    def inner(*args, **kwargs):
        print(func.__name__, end="...")
    
    return inner
```

```python
def custom_enumerate(data):
    counter = 0
    for element in data:
        counter += 1
        yield (counter, element)
```
