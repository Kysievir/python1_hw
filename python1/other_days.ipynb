```python
# Unhashables, or mutables, cannot be the key in a dictionary or an element in a set.
# This restriction allows "in" operation to be completed in O(1).
my_key = "key"
my_dict = {my_key: 50}
my_key += "_1"
print(my_key)
```

```python
my_dict.get("my_key", 0)
my_dict.update({"key2": 2})
my_dict.pop("key2")
```

```python
list(my_dict).sort() # in-place
list(my_dict).sorted()
```
